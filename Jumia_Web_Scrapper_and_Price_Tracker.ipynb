```python
from bs4 import BeautifulSoup
import requests 
import time 
import datetime


#connect to website 
source = requests.get('https://www.jumia.com.ng/fashion-mens-sports-t-shirtspants-suitwhite-94647986.html').text

soup = BeautifulSoup(source, 'lxml')
#print(soup.prettify())

_section = soup.find('section', class_ = 'col12 -df -d-co')
#print(_section.prettify())

title = _section.h1.text
print(title)

price_span = soup.find('span', {'class': '-b -ltr -tal -fs16'})
price = price_span.get_text(strip=True)
print(price) 
```

```python
import datetime

today = datetime.date.today()

print(today)
```

```python
import csv 

header = ['Title', 'Price', 'Date']
data = [title, price, today]


with open('jumia_shirt_scrapper.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
```
